## Helstrom's receipt

### Step 1 Build the matrix $\Gamma$

$$
\Gamma (\theta, P_{\Psi}, P_{\Phi}) = 
\begin{pmatrix}
(P_{\Psi}-P_{\Phi})cos^2(\theta) & \frac{1}{2}sin(2\theta)\\
\frac{1}{2}sin(2\theta)          & (P_{\Psi}-P_{\Phi})sin^2(\theta)
\end{pmatrix} 
$$

where:
* $P_{\Psi}$ : probability of state $\Psi$.
* $P_{\Phi}$ : probability of state $\Phi$.
* $\theta$ : the angle between both states (well in fact the angle between them in $2\theta$)

### Step 2 Diagonalize it

We will get two eigenvalues (one positive and one negative) and the corresponding eigenvectors that will be called:
* V_{+} : eigenvector associated to the positive eigenvalue
* V_{-} : eigenvector associated to the negative eigenvalue

### Step 3 Measure 

We bould our POVM with those eigenvectors:
* $E_{+} = |V_{+}><V_{+}|$
* $E_{-} = |V_{-}><V_{-}|$

Our estimator function is such that:
* If we measure $E{+}$ $\rightarrow$ state in $\Psi$
* If we measure $E{-}$ $\rightarrow$ state in $\Phi$

so to facilitate the notation let's define (more about the nightmare of notation later)
* $E_{\Psi} = |V_{+}><V_{+}|$
* $E_{\Phi} = |V_{-}><V_{-}|$

So we measure and based on the result ($E_{\Psi}$ or $E_{\Phi}$) we have a guess which is the state ($|\Psi>$ or $|\Phi>$).


## The nightmare of notation

Throw the papers the notation is not consistent. Let us put som examples to know we're talking abouy:
* States $\Psi>$ and $\Phi>$ are named as $0$ or $1$
* The measurements' operarators sometimes are $E_0$ or $\Pi_0$ 
* The values $\eta_{i}$ sometimes are the probablity of the experiment $H_{i}$ or the probability of the states $\Psi>$ and $\Phi>$, where in that case $\eta_0$ is the probability of the states $\Psi$
* An expression like $P(0|1)$ can be read as "probability of measuring the operator that my estimator says that is the state $|\Psi>$ when the states I have measured is $|\Phi>$
* And much more

So to try to avoid those confusions we will try to be very explicit even what that means the notation is *clumby*


## Bayesian + Helstrom's : a winner combination!

Now, how do we combine both things? 

### Step 1 Build the matrix $\Gamma$

In our case $\Gamma$ depends only of $P_{\Psi}$ and  $P_{\Phi}$ because the angle $\theta$ between the states is not going to change during the experiments, so the first thing is to compute those probabilities and that depends on:
* Probability of every experiment $P(H_{i})$
* Which particle of the 5 we are analyzing

For example let's suppose all the experiments have the same likelihood:
* $P(H_{1}) = P(H_{2}) = P(H_{3}) = P(H_{4}) = P(H_{5}) = \frac{1}{5}$

and we are analyzing the **first particle** where:
* State $|\Psi>$ is in the experiments $H_{2}, H_{3}, H_{4}, H_{5}$
* State $|\Phi>$ is in the experiment $H_{1}$

So applying basic probability 
* $P_{\Psi} = P(H_{2}) + P(H_{3}) + P(H_{4}) + P(H_{5}) = \frac{4}{5}$
* $P_{\Phi} = P(H_{1})                                  = \frac{1}{5}$


In general we have that when analyzing the **particle $s$** ($s \in [1,5]$)
* $P_{\Psi}^{s} = \sum_{j} P(H_{j})$
* $P_{\Phi}^{s} = \sum_{k} P(H_{k})$

where:
* j represents the experiments where the particle $s$ is $\Psi$
* k represents the experiments where the particle $s$ is $\Phi$

Just a last example to clarify : let's suppose we are measuring the third particle ($s=3$). In such case:
* $P_{\Psi}^{3} = P(H_{4}) + P(H_{5})$
* $P_{\Phi}^{3} = P(H_{1}) + P(H_{2}) + P(H_{3})$

So, we have to get those values that initialy (for $s=1$) will be (as we have shown in the previous example):
* $P_{\Psi}^{0} = \frac{4}{5}$
* $P_{\Phi}^{0} = \frac{1}{5}$

but those values will be updated when we perform some measurements becuase depending on the result some hipothesys will change its probability and that affects as we have seen when calculating $P_{\Psi}^{s}$ and $P_{\Phi}^{s}$. This "update of the priors" is done using the Bayes' rule but more about this later.

So, we already have the values of $P_{\Psi}$ and $P_{\Phi}$ so we can calculate $\Gamma$

### Step 2 Diagonalize it

So, this is "easy": diagonalize to get the the eigenvectors $V_{+}$ and $V_{-}$ so we can build:
* $E_{\Psi} = |V_{+}><V_{+}|$
* $E_{\Phi} = |V_{-}><V_{-}|$


### Step 3 Measure and update the priors

Ok, let's have some intuition here: let's suppose the particle we are analyzing is the third one ($s=3$) and in that case the states I can receive:
* $H_{5}^{3} = |\Psi>$
* $H_{4}^{3} = |\Psi>$
* $H_{3}^{3} = |\Phi>$
* $H_{2}^{3} = |\Phi>$
* $H_{1}^{3} = |\Phi>$


If we perform a measurement and we get $E_{\Psi}$ that means is more probable that the hypothesis are {$H_{5}$, $H_{4}$} than not the other three {$H_{3}$, $H_{2}$, $H_{1}$} and that means we will increase the values of $P(H_{5})$ and $P(H_{4})$ and decrease the others and we will take into consideration this change in the next particle $s=4$

This "update of the priors" (or update of the values $P(H_{i})$ is done using the Baye's rule. To do that we have to compute the following two values (acording our estimator):

* $P(E_{\Psi}|\Psi) = |<V_{+}|\Psi>|^2$
* $P(E_{\Phi}|\Phi) = |<V_{-}|\Phi>|^2$

Then depending on the hypotesis and which particle we are analyzing we "expect" a certain measurement. 

For example let's suppose we have measured the particle #3 ($s=3$) so based on the results we have the probabilities of every hypotesis and to to that we will use the Bayes' rule:

$$
P(A|B) = \frac {P(B|A)P(A)}{P(B)}
$$

where in our case:
* A : are the possible Hypotesis ($H_{i}$)
* B : are the results of the measurement ($E_{\Psi}$ and $E_{\Phi}$)

So something like:

$$
P(H_{|B) = \frac {P(B|A)P(A)}{P(B)}
$$

**Important remark** : as we have seen before the values of some of those quanties depend on $s$, the particle we're analyzing and for that reason we will use superindex to indicate that. With this on mind our Bayes's rules look like:

$$
P(H_{4}|E_{\Psi})^{3} = \frac {P(E_{\Psi}|H_{4})^{3} P(H_{4}^{2}) } { P(E_{\Psi})^{3} }
$$
