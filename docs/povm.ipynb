# Unambiguous binary state discrimination

**Given** the states:

$$
|\Psi_0> = \cos(\theta)|0> + \sin(\theta)|1> \\
|\Psi_1> = \cos(\theta)|0> - \sin(\theta)|1> \\
$$

where:
* $\theta \in [0,\pi]$
* $\eta_0$ : probability $|\Psi_0>$ to be chosen
* $\eta_1$ : probability $|\Psi_1>$ to be chosen

with probability $\eta_0$
* $|\Psi_1>$ is chosen with probability $\eta_1$

and the POVM:
$$
\pi_0 = \mu_0|\Psi_1^{\bot}><\Psi_1^{\bot}| \\
\pi_1 = \mu_1|\Psi_0^{\bot}><\Psi_0^{\bot}| \\
\pi_2 = \mathbb{1} - \pi_0 - \pi_1
$$

subject to the constraints:

$$
\mu_0 \geq 0 \\
\mu_1 \geq 0 \\
\pi_2 \geq 0
$$

**Proof** that the minimum probability of error, given by $P(\pi_2)$, is:

$$
P(\pi_2)^{min} = 2\sqrt{\eta_o\eta_1}<\Psi_0|\Psi_1> = \\
2\sqrt{\eta_o\eta_1}\cos(2\theta)
$$

## Demostration

### Some results

Some results that will be used:

$$
|\Psi_0^{\bot}> = \sin(\theta)|0> - \cos(\theta)|1> \\
|\Psi_1^{\bot}> = \sin(\theta)|0> + \cos(\theta)|1> \\
<\Psi_0^{\bot}|\Psi_1> = <\Psi_1^{\bot}|\Psi_0> = \sin(2\theta) \\
$$

where:
* $<\Psi_0^{\bot}|\Psi_0> = (\sin(\theta)<0| - \cos(\theta)<1|)(\cos(\theta)|0> + \sin(\theta)|1>)=\sin(\theta)\cos(\theta)<0|0>+\sin^2(\theta)<0|1>-\cos^2(\theta)<1|0>-\cos(\theta)\sin(\theta)<1|1> = 0$
* $<\Psi_1^{\bot}|\Psi_1> = (\sin(\theta)<0| + \cos(\theta)<1|)(\cos(\theta)|0> - \sin(\theta)|1>) = 0$
* $<\Psi_0^{\bot}|\Psi_1> = (\sin(\theta)<0| - \cos(\theta)<1|)(\cos(\theta)|0> - \sin(\theta)|1>) = \sin(\theta)\cos(\theta)<0|0> + \cos(\theta)\sin(\theta)<1|1> = 2\sin(\theta)cos(\theta)=\sin(2\theta)$
* $<\Psi_1^{\bot}|\Psi_0> = (\sin(\theta)<0| + \cos(\theta)<1|)(\cos(\theta)|0> + \sin(\theta)|1>) = 
\sin(\theta)\cos(\theta)<0|0> + \cos(\theta)\sin(\theta)<1|1> = 2\sin(\theta)cos(\theta)=\sin(2\theta)$



The probability distribution $P(\pi_i|\Psi_j)$:

|             |$\Psi_0$                 | $\Psi_1$               |
|---          |---                      |---                     |
|$\pi_0$      |$\mu_0\sin^2(2\theta)$   |$0$                     |
|$\pi_1$      |$0$                      |$\mu_1\sin^2(2\theta)$  |
|$\pi_2$      |$1-\mu_0\sin^2(2\theta)$ |$1-\mu_1\sin^2(2\theta)$ |


where:
* $P(\pi_0|\Psi_0) = tr (\pi_0|\Psi_0><\Psi_0|) = tr (\mu_0|\Psi_1^{\bot}><\Psi_1^{\bot}|\Psi_0><\Psi_0|) = \mu_0|<\Psi_1^{\bot}|\Psi_0>|^2 = \mu_0\sin^2(2\theta)$
* $P(\pi_1|\Psi_1) = tr (\pi_1|\Psi_1><\Psi_1|) = tr (\mu_1|\Psi_0^{\bot}><\Psi_0^{\bot}|\Psi_1><\Psi_1|) = \mu_1|<\Psi_0^{\bot}|\Psi_1>|^2 = \mu_1\sin^2(2\theta)$

### Calculate $P(\pi_2)$

$$
P(\pi_2) = 
P(\pi_2|\Psi_0)P(\Psi_0) + P(\pi_2|\Psi_1)P(\Psi_1) = \\
\eta_0(1-\mu_0\sin^2(2\theta)) + \eta_1(1-\mu_1\sin^2(2\theta)) = \\
\eta_0 + \eta_1 - (\eta_0\mu_0 + \eta_1\mu_1)\sin^2(2\theta) = \\
1 - (\eta_0\mu_0 + \eta_1\mu_1)\sin^2(2\theta) \equiv P_{\pi_2}(\mu_0,\mu_1)
$$ 

This is the quantity we want to minimize. We know the quantities $\eta_0$, $\eta_1$ and $\theta$ and the unknown are $\mu_0$ and $\mu_1$ and we want to find the values that minimize this function and this minimum value.

The next step is to try to find a relation between $\mu_0$ and $\mu_1$ using the constraints we know and specially the fact that $\pi_2 \ge 0$

### Use the fact $\pi_2 \ge 0$

The fact that $\pi_2 \ge 0$ means its eigenvalues are $\ge 0$

$$
\pi_2 = \mathbb{1} - \pi_0 - \pi_1 = \\
\mathbb{1} -  \mu_0|\Psi_1^{\bot}><\Psi_1^{\bot}| - \mu_1|\Psi_0^{\bot}><\Psi_0^{\bot}| = \\
\begin{pmatrix}
1 & 0 \\
0 & 1
\end{pmatrix} - 
\mu_0
\begin{pmatrix}
\sin(\theta) \\
\cos(\theta) \\
\end{pmatrix}
\begin{pmatrix}
\sin(\theta) & \cos(\theta)
\end{pmatrix} - 
\mu_1
\begin{pmatrix}
\sin(\theta) \\
-\cos(\theta) \\
\end{pmatrix}
\begin{pmatrix}
\sin(\theta) & -\cos(\theta)
\end{pmatrix} = \\
\begin{pmatrix}
1 & 0 \\
0 & 1
\end{pmatrix} - 
\mu_0
\begin{pmatrix}
\sin^2(\theta) & \sin(\theta)\cos(\theta) \\
\sin(\theta)\cos(\theta) & \cos^2(\theta) \\
\end{pmatrix} -
\mu_1
\begin{pmatrix}
\sin^2(\theta) & -\sin(\theta)\cos(\theta) \\
-\sin(\theta)\cos(\theta) & \cos^2(\theta) \\
\end{pmatrix} = \\
\begin{pmatrix}
1-(\mu_0+\mu_1)\sin^2(\theta) & (\mu_1-\mu_0)\sin(\theta)\cos(\theta) \\
(\mu_1-\mu_0)\sin(\theta)\cos(\theta) & 1-(\mu_0+\mu_1)\cos^2(\theta)
\end{pmatrix}
$$

If we compute the [eigenvalues](https://www.wolframalpha.com/input/?i=eigenvalues%7B%7B1-%28m0%2Bm1%29sin%5E2%28x%29%2C+%28m1-m0%29sin%28x%29cos%28x%29%7D%2C%7B%28m1-m0%29sin%28x%29cos%28x%29%2C+1-%28m0%2Bm1%29cos%5E2%28x%29%7D%7D):

$$
\lambda_1 = \frac{1}{2}\bigg( -\sqrt{\mu_0^2 + 2\mu_0\mu_1\cos(4\theta)+\mu_1^2} + 2 - (\mu_0 + \mu_1) \bigg) \ge 0 \\
\lambda_2 = \frac{1}{2}\bigg(  \sqrt{\mu_0^2 + 2\mu_0\mu_1\cos(4\theta)+\mu_1^2} + 2 - (\mu_0 + \mu_1) \bigg) \ge 0
$$

We can see easily the relation between $\mu_0$ and $\mu_1$ if we write the inequalities as

$$
-A + B \ge 0 \\
 A + B \ge 0 \\
$$

from what we easily deduce that the only possibility is

$$
A=B
$$

and that means in our case:

$$
\sqrt{\mu_0^2 + 2\mu_0\mu_1\cos(4\theta)+\mu_1^2} = 2 - (\mu_0 + \mu_1) \\
\mu_0^2 + 2\mu_0\mu_1\cos(4\theta)+\mu_1^2 = 4 + \mu_0^2 + \mu_1^2 + 2\mu_0\mu_1 - 4\mu_0 - 4\mu_1 \\
2\mu_0\mu_1\cos(4\theta) = 4 + 2\mu_0\mu_1 - 4\mu_0 - 4\mu_1 \\
\mu_0\mu_1\cos(4\theta) - \mu_0\mu_1 + 2\mu_0  = 2 - 2\mu_1 \\
\mu_0 ( \mu_1\cos(4\theta) - \mu_1 + 2) = 2 - 2\mu_1 \\
\mu_0 = \frac{2 - 2\mu_1 }{\mu_1\cos(4\theta) - \mu_1 + 2}
$$

Now let's replace this relation in our expression for $P_{\pi_2}(\mu_0,\mu_1)$ so we get an expression depending only of one variable:

$$
P_{\pi_2}(\mu_0,\mu_1) = 1 - ( \eta_0\mu_0 + \eta_1\mu_1 )\sin^2(2\theta) = \\
1 - \Bigg( \eta_0\bigg( \frac{2 - 2\mu_1 }{\mu_1\cos(4\theta) - \mu_1 + 2} \bigg)+ \eta_1\mu_1 \Bigg) \sin^2(2\theta) = \\
\Bigg(
  \frac
    {\mu_1\cos(4\theta) -\mu_1 +2 +2\eta_0 -2\eta_0\mu_1 +\eta_1\mu_1^2\cos(4\theta) -\eta_1\mu_1^2 + 2\eta_1\mu_1}
    {\mu_1\cos(4\theta) -\mu_1 +2}
\Bigg) 
\sin^2(2\theta) = P_{\pi_2}(\mu_1)
$$ 

So get get an expression that only depends of one parameter $\mu_1$. Now the nest steps is to find the minimum of this function with

$$
\frac{\partial{P_{\pi_2}(\mu_1)}}{\partial{\mu_1}} = 0
$$

Once we resolve this equation, we get a value for $\mu_1^{min}$ that we can verify if it is in fact a minimum using its second derivative. Once we get it we replace in $P_{\pi_2}$ to get the desired result:


$$
P(\pi_2)^{min} \equiv P_{\pi_2}(\mu_1^{min}) = 2\sqrt{\eta_o\eta_1}\cos(2\theta)
$$
