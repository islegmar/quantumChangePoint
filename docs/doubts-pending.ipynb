We have the states (computational basis)

$$
|\Psi_0> = \cos(\theta)|0>+\sin(\theta)|1> \\
|\Psi_1> = \cos(\theta)|0>-\sin(\theta)|1>
$$

and their corresponding density matrices:

$$
\rho_0\rvert_{computational} = |\Psi_0><\Psi_0| = 
\begin{pmatrix}
cos^2(\theta)            & \frac{1}{2}\sin(2\theta)\\
\frac{1}{2}\sin(2\theta) & sin^2(\theta)
\end{pmatrix} \\
\rho_1\rvert_{computational} = |\Psi_1><\Psi_1| = 
\begin{pmatrix}
cos^2(\theta)             & -\frac{1}{2}\sin(2\theta)\\
-\frac{1}{2}\sin(2\theta) & sin^2(\theta)
\end{pmatrix}
$$


Measuring using Helstrom means using the following operators that have this representation in the eigenbasis of $\Gamma$: 

$$
E_0\rvert_{\Gamma} = \begin{pmatrix} 1 &0 \\ 0 & 0 \end{pmatrix} \\
E_1\rvert_{\Gamma} = \begin{pmatrix} 1 &0 \\ 0 & 1 \end{pmatrix} \\
$$

where $\Gamma$ in the computationa basis looks like:


$$
\Gamma(\theta, \eta_0, \eta_1)\rvert_{computational} = 
\eta_0\rho_0 - \eta_1\rho_1 = 
\begin{pmatrix}
(\eta_0-\eta_1)cos^2(\theta) & \frac{1}{2}sin(2\theta)\\
\frac{1}{2}sin(2\theta)      & (\eta_0-\eta_1)sin^2(\theta)
\end{pmatrix}
$$

Ok, now in our experiment the **first particle** can be $\Psi_0$ or $\Psi_1$ with the following priors:
* $\eta_0 = \frac{4}{5}$ of being $|\Psi_0>$
* $\eta_1 = \frac{1}{5}$ of being $|\Psi_1>$

and that means:

$$
\Gamma(\theta, \eta_0=\frac{4}{5}, \eta_1=\frac{1}{5}) = 
\eta_0\rho_0 - \eta_1\rho_1 = 
\begin{pmatrix}
\frac{3}{5}cos^2(\theta) & \frac{1}{2}sin(2\theta)\\
\frac{1}{2}sin(2\theta)      & \frac{3}{5}sin^2(\theta)
\end{pmatrix}
$$


Ok, now we want to calculate the new priors and that means the probability of measuring $E_0$ and $E_1$ in the mixed state:

$$
\rho = \eta_0\rho_0 + \eta_1\rho_1
$$ 

The probabbility of measuring $E_0$ is given for:

$$
P(E_0|\rho) = tr(E_0\rho) = tr(E_0(\eta_0\rho_0 + \eta_1\rho_1)) = \\
tr(E_0(\eta_0\rho_0 + \eta_1\rho_1 + \eta_1\rho_1 - \eta_1\rho_1)) = \\
tr(E_0(\Gamma + 2\eta_1\rho_1)) = tr(E_0\Gamma) + 2\eta_1tr(E_0\rho_1)
$$

Ok, let's compute both terms:

$\underline{tr(E_0\Gamma)}$ : becuse the trace is invariant under chabnge of basis, if we express $E_0$ and $\Gamma$ in the eigenbasis of $\Gamma$ the output is trivial:

$$
tr(E_0\Gamma)\rvert_{\Gamma} = 
tr \bigg[ 
  \begin{pmatrix} 1 &0 \\ 0 & 0 \end{pmatrix} 
  \begin{pmatrix} \lambda_1 &0 \\ 0 & \lambda_2 \end{pmatrix}
\bigg] =
\lambda_1
$$

$\underline{tr(E_0\rho_1)}$ : ok, here is when probably I'm missing something and I'm overcomplicating the things or ... this is how it is

To calculate it we must have both therms ($E_0$ and $\rho_1$) in the same basis that can be:

* **Computational** : $\rho_1\rvert_{computational} =  
\begin{pmatrix}
cos^2(\theta)             & -\frac{1}{2}\sin(2\theta)\\
-\frac{1}{2}\sin(2\theta) & sin^2(\theta)
\end{pmatrix}$ and $E_0\rvert_{computational}$ unknown

* **$\Gamma$'s eigenbasis** : $\rho_1\rvert_{\Gamma}$ unknown and $E_0\rvert_{\Gamma} = \begin{pmatrix} 1 &0 \\ 0 & 0 \end{pmatrix}$ 


where we have the relations:

$$
V^{\dagger} E_0\rvert_{computational} V = E_0\rvert_{\Gamma} = \begin{pmatrix} 1 &0 \\ 0 & 0 \end{pmatrix} \\
E_0\rvert_{computational} = V E_0\rvert_{\Gamma} V^{\dagger}= V \begin{pmatrix} 1 &0 \\ 0 & 0 \end{pmatrix} V^{\dagger}
$$


where $V$ is the matrix with the eigenvectors of $\Gamma$ that in general al "quite complex"; for example in our case 

$$
\Gamma(\theta, \eta_0=\frac{4}{5}, \eta_1=\frac{1}{5}) = 
\eta_0\rho_0 - \eta_1\rho_1 = 
\begin{pmatrix}
\frac{3}{5}cos^2(\theta) & \frac{1}{2}sin(2\theta)\\
\frac{1}{2}sin(2\theta)      & \frac{3}{5}sin^2(\theta)
\end{pmatrix}
$$


we get that the [eigenvectors are](https://www.wolframalpha.com/input/?i=eigenvalues+%7B%7B3%2F5cos%5E2%28x%29%2C+1%2F2sin%282x%29%7D%2C+%7B1%2F2sin%282x%29%2C+3%2F5sin%5E2%28x%29%7D%7D)

So now the steps should be:
* Build $V$ (the eigenvectors as columns)
* Compute the result of $tr(E_0\rho_1)\rvert_{computational}$ that would like something like:

$$
tr(E_0\rho_1)\rvert_{computational} = 
tr(E_0\rvert_{computational}\rho_1\rvert_{computational}) = \\
tr\bigg[
  V 
  \begin{pmatrix} 
    1 &0 \\ 
    0 & 0 
  \end{pmatrix} 
  V^{\dagger}
  \begin{pmatrix}
    cos^2(\theta)             & -\frac{1}{2}\sin(2\theta)\\
    -\frac{1}{2}\sin(2\theta) & sin^2(\theta)
  \end{pmatrix}
\bigg]
$$

## $\theta=\frac{\pi}{8}$

Just to check we can found the solution in [Wolphram Alpha](https://www.wolframalpha.com/input/?i=eigenvalues+%7B%7B3%2F5cos%5E2%28pi%2F8%29%2C+1%2F2sin%282pi%2F8%29%7D%2C+%7B1%2F2sin%282pi%2F8%29%2C+3%2F5sin%5E2%28pi%2F8%29%7D%7D)


```python
import numpy as np
import math
from numpy import linalg as LA

E0=np.array([[1,0],[0,0]])
E1=np.array([[0,0],[0,1]])

def compute(theta, etha0, etha1):
    """ Compute the new probabilities for E0 and E1"""
    
    rho0=np.array([
        [math.cos(theta)**2, (1/2)*math.sin(2*theta)],
       [(1/2)*math.sin(2*theta), math.sin(theta)**2]
    ])
    rho1=np.array([
        [math.cos(theta)**2      , (-1/2)*math.sin(2*theta)],
        [(-1/2)*math.sin(2*theta), math.sin(theta)**2]
    ])

    gamma = etha0*rho0 - etha1*rho1
  
    W, V = LA.eig(gamma)
 
    return np.trace(V * E0 * np.matrix.getH(V) * rho1)


print (compute(np.pi/8, 4/5, 1/5))
```
